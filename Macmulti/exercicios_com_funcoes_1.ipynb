## Exercício 1
Um número $a$ é dito permutação de um número $b$ se os dígitos de $a$ formam uma permutação dos dígitos de $b$.

Exemplo:   $5412434$ é uma permutação de $4321445$, mas não é uma permutação de $4312455$.

Obs.: Considere que o dígito $0$ (zero) não aparece nos números.

(a) Faça uma função contadígitos que dados um inteiro $n$ e um inteiro $d$, $0 < d < 9$, devolve quantas vezes o dígito $d$ aparece em $n$.

(b) Usando a função do item anterior, faça um programa que lê dois inteiros positivos $a$ e $b$ e responda se $a$ é permutação de $b$.

```python
def contadigitos(n, d):
    """Função contadígitos, recebe um inteiro n e um inteiro d"""
    n_str = str(n)
    d_str = str(d)
    cont = 0
    
    for i in range(len(n_str)):
        if n_str[i] == d_str:
            cont += 1
    return cont


def is_permutacao(a, b):
    """Função que recebe dois números e verifica se são permutação"""
    a_str = str(a)
    b_str = str(b)
    flag = True
    
    if len(a_str) == len(b_str):
        for i in range(10):
            if contadigitos(a_str, i) != contadigitos(b_str, i):
                flag = False
        
        return flag        
    else:
        return False
    
    
def main():
    a = int(input('a: '))
    b = int(input('b: '))
    
    if is_permutacao(a, b):
        print('É permutação')
    else:
        print('Não é permutação')

        
main()
```

(a) Construa uma função encaixa que dados dois inteiros positivos $a$ e $b$ verifica se $b$ corresponde aos últimos dígitos de $a$.

Ex.:


|    a   |   b  |                |
|:------:|:----:|:--------------:|
| 567890 |  890 |   => Encaixa   |
|  1243  | 1243 |   => Encaixa   |
|  2457  |  245 | => Não encaixa |
|   457  | 2457 | => Não encaixa |

(b) Usando a função do item anterior, faça um programa que lê dois inteiros positivos $a$ e $b$ e verifica se o menor deles é segmento do outro.

|    a   |    b    |                               |
|:------:|:-------:|:-----------------------------:|
| 567890 |   678   |      => b é segmento de a     |
|  1243  | 2212435 |      => a é segmento de a     |
|   235  |   236   | => um não é segmento do outro |

```python
def encaixa(a, b):
    """Função que recebe a e b, e verifica se b encaixa nos ultimos digitos de a"""
    while b != 0 and (a % 10 == b % 10):
        a = a // 10
        b = b // 10

    if b == 0:
        return True
    else:
        return False

    
def main():
    a = int(input('a: '))
    b = int(input('b: '))
    flag = False
    
    if a > b:
        maior = a
        menor = b
    else:
        maior = b
        menor = a
        
    maior_p, menor_p = maior, menor  # só para printar bonitinho
    
    while maior >= menor:
        if encaixa(maior, menor):
            flag = True
        maior = maior // 10
    
    if flag:
        print(f'{menor_p} é segmento de {maior_p}')
    else:
        print('um não é segmento do outro')
    
main()
```

# Exercício 3
*(MAT - 94)* Uma seqüência de $n$ números inteiros não nulos é dita piramidal $m$-alternante se é constituída por $m$ segmentos: o primeiro com um elemento, o segundo com dois elementos e assim por diante até o $m$-ésimo, com $m$ elementos. Além disso, os elementos de um mesmo segmento devem ser todos pares ou todos ímpares e para cada segmento, se seus elementos forem todos pares (ímpares), os elementos do segmento seguinte devem ser todos ímpares (pares).

Por exemplo, 

a seqüência com $n = 10$ elementos:

<ins>12</ins> <ins>3  7</ins>  <ins>2  10  4</ins> <ins>5  13  5  11</ins> é piramidal $4$-alternante.

A seqüência com $n = 3$ elementos:

<ins>7</ins> <ins>10  2</ins> é piramidal $2$-alternante.

A seqüência com $n = 8$ elementos:

<ins>1</ins> <ins>12 4</ins> <ins>3  13  5</ins> <ins>12  6</ins> não é piramidal alternante pois o último segmento não tem tamanho $4$.

(a) Escreva uma função *bloco* que recebe como parâmetro um inteiro $n$ e lê n inteiros do teclado, devolvendo um dos seguintes valores:

* 0,  se os $n$ números lidos forem pares;
* 1,  se os $n$ números lidos forem ímpares;
* -1,  se entre os $n$ números lidos há números com paridades diferentes.

(b) usando a função do item anterior, escreva um programa que, dados um inteiro $n > 1$ e uma seqüência de $n$ números inteiros, verifica se ela é piramidal $m$-alternante. O programa deve imprimir o valor de $m$ ou dar a resposta não

```python
def bloco(n, element = None):
    if not element:
        elementos = [int(input(f'{i+1}º elemento: ')) for i in range(n)]
    else:
        elementos = element
        
    par = False
    impar = False
    
    for ele in elementos:
        if ele % 2 == 0:
            par = True
        else:
            impar = True
            
    if par and impar:
        return -1
    else:
        if par:
            return 0
        if impar:
            return 1

        
def main():
    n = int(input('n: '))
    lista_elementos = [int(input(f'{i+1}º elemento: ')) for i in range(n)]
    alt = True
    
    # só para achar a sequencia de que vai de 1, 3, 6, 10, 15, 21...
    soma = 0
    lista_seq = []
    for i in range(1,n+1):
        soma+=i
        lista_seq.append(soma)
    
    if n in lista_seq:
        m = lista_seq.index(n) + 1  # números de cortes
        elementos_div = []
        i = 1
        for j in range(m):
            elementos_div.append(lista_elementos[j:j+i])
            i+=1
        
        for i in elementos_div:
            if bloco(len(i),i) == -1:
                alt = False
        
        if alt:
            print(f'É {m}-alternante')
        else:
            print('Não é alternante')
        
    else:
        print('Não é alternante')

main()
```

(c) Faça um programa que, dados $n$ pontos do primeiro quadrante $(x > 0$ e $y > 0)$ através de suas coordenadas cartesianas, determina o ponto que forma o menor ângulo com o eixo horizontal. Use a função do item anterior, mesmo que você não a tenha feito.


```python
def arctan(x):
    k = 1
    soma = x
    while abs((x**k) / k) > 0.0001:
        soma += (((-1) ** k) / (2*k + 1)) * (x ** (2*k + 1))
        k+=1
        
    return soma

def angulo(x, y):
    pi = 3.1415926535
    if y < x:
        ang = arctan(y/x)
    else:
        ang = pi/2 - arctan(x/y)
    
    return ang*180/pi
    
def main():
    n = int(input('n: '))
    x_m = 0
    y_m = 0
    val_m = 361  # valor de sacrificio

    for i in range(n):
        x = float(input('x: '))
        y = float(input('y: '))
        
        result = angulo(x, y)
        
        if result < val_m:
            val_m = result
            x_m = x
            y_m = y
    
    print(f'O menor angulo foi {val_m} graus, do ponto ({x_m},{y_m})')
        
main()
```

## Exercício 5
(a) Escreva uma função que recebe um inteiro positivo $m$ e devolve $1$ se $m$ é primo, $0$ em caso contrário.

(b) Escreva um programa que leia um inteiro não-negativo $n$ e imprima a soma dos $n$ primeiros números primos.

```python
def is_primo(m):
    soma = 0
    
    for cand in range(1, m+1):
        if m%cand == 0:
            soma +=1
            
    if soma > 2:
        return 0  # não é primo
    else:
        return 1  # é primo
    
def main():
    n = int(input('n: '))
    k = 0
    num = 1
    while k < n:
        if is_primo(num) == 1:
            print(num, end=' ')
            k += 1
        num += 1
        
    
main()
```

## Exercícios 6
(a) Escreva uma função que recebe como parâmetros dois números $a$ e $b$ e devolve o *mdc* (máximo divisor comum) de $a$ e $b$, calculado por meio do algoritmo de Euclides.

(b) Escreva um programa que leia um inteiro positivo $n$ e uma seqüência de $n$ inteiros não-negativos e imprime o *mdc* de todos os números da seqüência.

```python
def mdc(a, b):

    # algoritmo de Euclides
    anterior = a
    atual = b
    resto = anterior % atual

    while resto != 0:
        anterior = atual
        atual    = resto
        resto    = anterior % atual

    return atual

def main():
    n = int(input('n: '))
    seq = [int(input(f'{i+1}')) for i in range(n)]
    
    if n >= 2
        atual = mdc(seq[0], seq[1])
        for i in range(2, n):
            novo_atual = mdc(atual, seq[i])
            atual = novo_atual
    
        print(atual)
    else:
        print('Digite pelo menos uma sequência com 2 ou mais números')
    
main()
```
