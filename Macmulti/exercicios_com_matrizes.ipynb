## Exercício 1
Dada uma matriz real $A$ com $m$ linhas e $n$ colunas e um vetor real $V$ com $n$ elementos, determinar o produto de $A$ por $V$.

```python
# dei uma enfeitada em algumas partes!
def main():
    m = int(input('Digite o número de linhas da matriz: '))
    n = int(input('Digite o número de colunas da matriz: '))
    
    matriz = []
    vetor = []
    m_v = []
    
    for i in range(m):
        linha = []
        for j in range(n):
            x = float(input(f'Elemento a {i+1}x{j+1}: '))
            linha.append(x)
            
        matriz.append(linha)
    
    for i in range(n):
        x = float(input(f'Digite o {i+1}º elemento do vetor V:'))
        vetor.append(x)
        
    print('Matriz A:')
    for i in range(m):
        print(matriz[i])
    
    print('Vetor V:')
    print(vetor)
    
    print('A x V:')
    for i in range(m):
        x = 0
        for j in range(n):
            x += vetor[j]*matriz[i][j]
            
        m_v.append([x])
        
    for i in range(m):
        print(m_v[i])
     
        
main()
```

## Exercício 2
Um vetor real $X$ com $n$ elementos é apresentado como resultado de um sistema de equações lineares $Ax = B$ cujos coeficientes são representados em uma matriz real $A_{mxn}$ e os lados direitos das equações em um vetor real $B$ de $m$ elementos. Verificar se o vetor $X$ é realmente solução do sistema dado.


```python
def main():
    m = int(input('Digite o número de linhas da matriz A (m): '))
    n = int(input('Digite o número de colunas da matriz A (n): '))
    
    matriz_a = []
    for i in range(m):
        linha = []
        for j in range(n):
            x = float(input(f'Elemento a {i+1}x{j+1}: '))
            linha.append(x)
        matriz_a.append(linha)
    
    vetor_b = []
    for i in range(m):
        x = float(input(f'Digite o {i+1}º elemento do vetor B: '))
        vetor_b.append(x)
    
    vetor_x = []
    for i in range(n):
        x = float(input(f'Digite o {i+1}º elemento do vetor X: '))
        vetor_x.append(x)
    
    resultado = []
    for i in range(m):
        soma = 0
        for j in range(n):
            soma += matriz_a[i][j] * vetor_x[j]
        resultado.append(soma)
    
    eh_solucao = True
    for i in range(m):
        if abs(resultado[i] - vetor_b[i]) > 1e-10:
            eh_solucao = False
            break
    
    print('Matriz A:')
    for i in range(m):
        print(matriz_a[i])
    
    print('Vetor B:')
    print(vetor_b)
    
    print('Vetor X:')
    print(vetor_x)
    
    print('A*X:')
    print(resultado)
    
    if eh_solucao:
        print('O vetor X é solução do sistema Ax = B')
    else:
        print('O vetor X NÃO é solução do sistema Ax = B')

main()
```

## Exercício 3
Dadas duas matrizes reais  $A_{mxn}$ e $B_{nxp}$, calcular o produto de $A$ por $B$.

```python
def main():
    m = int(input('Digite o m da matriz A m x n: '))
    n = int(input(f'Digite o n da matriz A {m} x n e B n x p: '))
    p = int(input(f'Digite o p da matriz B {n} x p: '))
    
    matriz_a = []
    matriz_b = []
    prod_matriz = []
    
    for i in range(m):
        linha = []
        for j in range(n):
            x = float(input(f'Insira o elemento a {i+1}x{j+1}: '))
            linha.append(x)
            
        matriz_a.append(linha)
        
    for i in range(n):
        linha = []
        for j in range(p):
            x = float(input(f'Insira o elemento b {i+1}x{j+1}: '))
            linha.append(x)
            
        matriz_b.append(linha)
        
    for i in range(m):
        linha = []
        for j in range(p):
            soma = 0
            for k in range(n):
                soma += matriz_a[i][k] * matriz_b[k][j]
            linha.append(soma)
        
        prod_matriz.append(linha)
            
    for i in range(m):
        print(prod_matriz[i])
    
    
main()
```

## Exercício 4
Dada uma matriz real  $A_{mxn}$, verificar se existem elementos repetidos em $A$.

```python
def main():
    m = int(input('Digite o número de linhas da matriz: '))
    n = int(input('Digite o número de colunas da matriz: '))
    
    matriz = []
    achou = False
    
    for i in range(m):
        linha = []
        for j in range(n):
            x = float(input(f'Elemento a {i+1}x{j+1}: '))
            linha.append(x)
            
        matriz.append(linha)

    for i in range(m):
        for j in range(n):
            valor_teste = matriz[i][j]
            for k in range(m):
                for v in range(n):
                    if valor_teste == matriz[k][v] and (i != k or j !=v):
                            achou = True
            
    if achou:
        print('Existem elementos repetidos')
    else:
        print('Não existem elementos repetidos')
    
    
main()
```

## Exercício 6
Dada uma matriz  $A_{mxn}$, imprimir o número de linhas e o número de colunas nulas da matriz.

Exemplo: $m = 4$ e $n = 4$

<center><img src='https://www.ime.usp.br/~macmulti/figuras/Image66.gif'></center>

tem $2$ linhas nulas e $1$ coluna nula.

```python
def main():
    m = int(input('Digite o número de linhas da matriz: '))
    n = int(input('Digite o número de colunas da matriz: '))
    
    matriz = []
    n_l = 0
    n_c = 0
    
    for i in range(m):
        linha = []
        for j in range(n):
            x = float(input(f'Elemento a {i+1}x{j+1}: '))
            linha.append(x)
            
        matriz.append(linha)
        
    for i in range(m):
        linha = matriz[i]
        cont_l = 0
        cont_c = 0
        for j in range(n):
            if matriz[j][i] == 0:
                cont_c += 1
            
            if linha[j] == 0:
                cont_l += 1
        
        if cont_l == n:
            n_l += 1
        if cont_c == m:
            n_c += 1
            
    print(f'Linhas nulas: {n_l}')
    print(f'Colunas nulas: {n_c}')    
        
    
main()
```

## Exercício 7
Dizemos que uma matriz quadrada inteira é um *quadrado mágico* se a soma dos elementos de cada linha, a soma dos elementos de cada coluna e a soma dos elementos das diagonais principal e secundária são todas iguais.

Exemplo: A matriz

<center><img src='https://www.ime.usp.br/~macmulti/figuras/Image67.gif'></center>

é um *quadrado mágico*.

Dada uma matriz quadrada $A_{nxn}$ , verificar se $A$ é um *quadrado mágico*.

```python
def main():
    m = int(input('Digite a ordem da matriz: '))

    matriz = []
        
    flag = True

    s_diag_p = 0
    s_diag_s = 0
    
    for i in range(m):
        linha = []
        for j in range(m):
            x = float(input(f'Elemento a {i+1}x{j+1}: '))
            linha.append(x)
            
        matriz.append(linha)
   
    for i in range(m):
        soma_col = 0
        soma_linha = 0
        
        s_diag_s += matriz[m-i-1][i]
        s_diag_p += matriz[i][i]
        for j in range(m):
            soma_linha += matriz[i][j]
            soma_col += matriz[j][i]      
            
        if soma_linha != soma_col:
            flag = False
    
    if s_diag_p != s_diag_s:
        flag = False
    
    if flag:
        print('É um quadrado mágico')
    else:
        print('Não é um quadrado mágico')
        
        
main()
```

## Exercício 9
(MAT 89) Um jogo de palavras cruzadas pode ser representado por uma matriz  $A_{mxn}$  onde cada posição da matriz corresponde a um quadrado do jogo, sendo que $0$ indica um quadrado branco e $-1$ indica um quadrado preto. Indicar na matriz as posições que são início de palavras horizontais e/ou verticais nos quadrados correspondentes (substituindo os zeros), considerando que uma palavra deve ter pelo menos duas letras. Para isso, numere consecutivamente tais posições.

Exemplo: Dada a matriz

<center><img src='https://www.ime.usp.br/~macmulti/figuras/Image69a.gif'></center>

A saída deverá ser

<center><img src='https://www.ime.usp.br/~macmulti/figuras/Image69b.gif'></center>

```python
def main():
    m = int(input('m: '))
    n = int(input('n: '))
    
    A = [[int(input(f'Digite o valor A_{i}_{j}')) for j in range(n)] for i in range(m)]
    B = [row[:] for row in A] 
    
    contador = 1
  
    for i in range(m):
        for j in range(n):
            if A[i][j] != 0:
                continue

            horiz = (j+1 < n and A[i][j+1] == 0 and (j-1 < 0 or A[i][j-1] == -1))
            vert = (i+1 < m and A[i+1][j] == 0 and (i-1 < 0 or A[i-1][j] == -1))

            if horiz or vert:
                B[i][j] = contador
                contador += 1

    print('Matriz A:')
    for i in range(m):
        for j in range(n):
            print(f'{A[i][j]:>3}', end=' ')
        print()
    
    print('Matriz B:')
    for i in range(m):
        for j in range(n):
            print(f'{B[i][j]:>3}', end=' ')
        print()

main()
```
