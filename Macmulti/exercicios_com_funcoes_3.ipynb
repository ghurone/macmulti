NOTE que não é preciso ler todos os conjuntos de uma só vez. Você pode ler os dois primeiros conjuntos e calcular a primeira intersecção. Depois, leia o próximo conjunto e calcule uma nova intesecção entre esse conjunto lido e o conjunto da intersecção anterior, e assim por diante.

Use obrigatoriamente a  função do item anterior, mesmo que você não a tenha feito.

```python
# acho que faz mais sentido fazer esse exercício esquecendo que a função len() existe. E que o [0] tem o tamanho do conjuto

def intersec(A:list[int] , B:list[int]) -> list[int]:
    """
    (list A, list B) -> list C
    """
    # Caso houver elementos repetidos nos conjuntos A e B, remove-los
    # Removendo os elementos repetidos em A
    tam_a = A[0]
    temp = []
    for i in range(1, tam_a):
        if A[i] not in temp:
            temp.append(A[i])
        else:
            A[0] -= 1
    A[1:-1] = temp[:]
    
    # Removendo os elementos repetidos em B
    tam_b = B[0]
    temp = []
    for i in range(1, tam_b):
        if B[i] not in temp:
            temp.append(B[i])
        else:
            B[0] -= 1
    B[1:-1] = temp[:]
    
    C = [0]  #lista da intersecao
    tam_c = 0
    for i in range(1, A[0]+1):
        for j in range(1, B[0]+1):
            if A[i] == B[j]:
                C.append(A[i])
                tam_c += 1
                C[0] = tam_c
    
    return C
                
def main():
    n = int(input('n: '))
    
    all_seq = []
    for i in range(n):
        seq = [0]
        while True:
            x = int(input(f'Digite um número para a {i+1}ª sequência, 0 para finalizar: '))
            if x == 0:
                break
            else:
                seq.append(x)
                seq[0] += 1
        all_seq.append(seq)
        
    for i in range(n-1):
        INTER = intersec(all_seq[i], all_seq[i+1])
        all_seq[i+1] = INTER
    
    print(INTER)
    
main()
```

(a) Escreva uma função que lê, linha a linha, uma matriz real $A_{mxn}$

## Exercício 3
(a) Escreva uma função que calcula a soma dos elementos da linha $i$ de uma matriz real $A_{mxn}$.

(b) Escreva uma função que calcula o produto dos elementos da coluna $j$ de uma matriz real $A_{mxn}$.

```python
def soma_elementos_linha(matriz:list[list[float]], linha:int) -> float:
    soma = 0
    for num in matriz[linha]:
        soma += num
        
    return soma

def prod_elementos_coluna(matriz:list[list[float]], coluna:int) -> float:
    soma = 1
    for n in len(matriz[0]):
        soma *= matriz[n][coluna]
        
    return soma
```

## Exercício 7

a) Escreva uma função que recebe como parâmetros um vetor real $A$ com $n$ elementos e um vetor real $B$ com m elementos, ambos representando conjuntos, e verifica se $A$ está contido em $B$ ($A \subseteq B$)

b) Usando a função do item acima verifique se dois conjuntos são iguais ($A = B$ se e somente se  $A \subseteq B$ e $B \subseteq A$) 

```python
def contido(A:list, B:list) -> bool:
    for ele in A:
        if ele not in B:
            return False
    
    return True

def igual(A:list, B:list) -> bool:
    return contido(A, B) and contido(B, A)
```

# Exercício 9

(a) Dado um vetor real $X$ com $n$ elementos e um certo índice $k$, escreva uma função que determina o índice do elemento mínimo entre $X[k], X[k+1], ..., X[n-1]$.

(b) Usando a função do item anterior coloque os elementos de um vetor em ordem crescente

```python
def menor_i(vetor:list[float], k:int) ->int:
    ind = k
    menor = vetor[ind]
    for i in range(k, len(vetor)):
        if menor > vetor[i]:
            menor = vetor[i]
            ind = i
    
    return ind

def ordenar(vetor:list[float]) -> None:
    for i in range(len(vetor)):
        ind = menor_i(vetor, i)
        vetor[i], vetor[ind] = vetor[ind], vetor[i]
```

# Exercício 10

Para encontrar uma raiz de um polinômio $P(x)=a_0 + a_{1}x_{1}+...+a_{n}x_{n}$, $(n \geq 2)$, pode-se aplicar o método de Newton, que consiste em refinar uma aproximação inicial $x_0$ dessa raiz através da expressão:

$x_{n+1} = x_{n} - \frac{P(x_{n})}{P'(x_{n})}$

$n = 0, 1, 2, ...$,  onde $P'(x)$ é a primeira derivada de $P(x)$.


Usualmente, repete-se esse refinamento até que $|x_{n+1} - x{n}| < \epsilon$ , $\epsilon \geq 0$, ou até que $m$ iterações tenham sido executadas.

Dados um polinômio $P(x)=a_0 + a_{1}x_{1}+...+a_{n}x_{n}$, uma aproximação inicial $x_0$ da raiz de $P(x)$, $\epsilon \geq 0$ e o número máximo de iterações que devem ser executadas, determine uma aproximação da raiz de $P(x)$ pelo método de Newton. Utilize uma função que, dado um polinômio $P(x)$, calcula a derivada $P'(x)$ e, para calcular $P(x_n)$ e $P'(x_n)$ em cada iteração, uma função que calcula o valor de um polinômio em um ponto.

```python
# Se preferir pode utilizar uma lambda function para definir o polinömio

def pol(coefs:list[float | int], x:float | int ) -> float:
    result = 0
    for i in range(len(coefs)):
        result += coefs[i] * x ** i
    
    return result

def derivada(coefs:list[float | int], x:float | int ) -> float:
    coef_d = []
    for i in range(1, len(coefs)):
        coef_d.append(coefs[i]*i)

    return pol(coef_d, x)

def newton(coefs:list[float | int], x0:float, m:int=20, eps:float=0.0001):
    for _ in range(m):
        x_n = x0 - pol(coefs, x0) / derivada(coefs, x0)
        
        if abs(x_n - x0) < eps:
            break
        
        x0 = x_n
    
    return x_n
```

```python
# Exercício (a) (ordenando com o bubblesort)

def ord_list(A:list[list[int]], c:int, k:int, p:int) -> None:
    for i in range(p, k-1, -1):
        for j in range(i):
            if A[j][c] > A[j+1][c]:
                A[j], A[j+1] = A[j+1], A[j]

# Exercício (b)
def main():
    n = int(input('Digite a quantidade de datas: '))
    
    DATA = []
    for i in range(n):
        data = input(f'Digite a {i+1}ª data, separando o dia, mês e ano por um espaço: ').split(' ')
        data = [int(i) for i in data]  # Converte para inteiro
        DATA.append(data)
    
    ord_list(DATA,2,0,len(DATA)-1)

main()
```
