## Exercício 1
(a) Escreva uma função que recebe um número inteiro $n>0$ e devolve o número de dígitos de $n$ e o primeiro dígito de $n$.

(b) Escreva um programa que leia uma seqüência de $n$ inteiros positivos e imprime o número de dígitos e o primeiro dígito de cada um deles.

````python
def func_lol(n:int) -> tuple[int]:
    n_str = str(n)
    return (len(n_str), int(n_str[0]))

def main():
    n = int(input('n: '))
    seq = [int(input(f'{i+1}º: ')) for i in range(n)]
    
    for num in seq:
        a, b = func_lol(num)
        print(f'{num} - Número de dígitos: {a} - Primeiro dígito: {b}')
    
main()
````

## Exercício 2

(a) Escreva uma função que recebe como parâmetro um inteiro positivo ano e devolve $1$ se ano for bissexto, $0$ em caso contrário. Um ano é bissexto se 
```C
(ano % 4 == 0 && (ano % 100 != 0 || ano % 400 == 0))
```
(b) Escreva uma função que tem como parâmetros de entrada e saída três números inteiros, dia, mes e ano, representando uma data, e modifica esses inteiros de forma que eles representem o dia seguinte.

(c) Escreva um programa que leia um inteiro positivo $n$ e uma seqüência de $n$ datas e imprime, para cada data, o dia seguinte.

````python
def is_bi(ano: int) -> int:
    if ano % 4 == 0 and (ano % 100 != 0  or ano % 400 == 0):
        return 1
    else:
        return 0

def dia_seguinte(dia:int, mes:int, ano:int) -> tuple[int]:
    meses = {'1': 31, '2': 28, '3': 31, '4': 30,
             '5': 31, '6': 30, '7': 31, '8': 31,
             '9': 30, '10': 31, '11': 30, '12': 31}
    
    if is_bi(ano) == 1:
        meses['2'] = 29
    
    if meses[str(mes)] == dia:
        dia = 1
        if mes == 12:
            mes = 1
            ano += 1
        else:
            mes+=1
    else:
        dia += 1
    
    return dia, mes, ano

def main():
    n = int(input('n: '))
    l_dia, l_mes, l_ano = [], [], []
    
    for i in range(n):
        l_dia.append(int(input('dia: ')))
        l_mes.append(int(input('mês: ')))
        l_ano.append(int(input('ano: ')))
    
    for j in range(n):
        dia, mes, ano = dia_seguinte(l_dia[j], l_mes[j], l_ano[j])
        print(f'{dia}/{mes}/{ano}')
    
main()
````

## Exercício 3
(a) Escreva uma função de cabeçalho 
```C
int divide (int *m, int *n, int d)
```
que recebe três inteiros positivos como parâmetros e devolve $1$ se $d$ divide pelo menos um entre $*m$ e $*n$, $0$ caso contrário. Fora isso, se $d$ divide $*m$, divide $*m$ por $d$, e o mesmo para o $*n$.

(b) Escreva um programa que lê dois inteiros positivos m e n e calcula, usando a função acima, o mínimo múltiplo comum entre m e n, ou seja, mmc(m,n).

````python
def divide(m:int, n:int, d:int) -> int:
    if d % m == 0 and d % n == 0:
        return 1
    else:
        return 0
    
def main():
    m = int(input('m: '))
    n = int(input('n: '))
    
    if m > n:
        maior = m
        menor = n
    else:
        maior = n
        menor = m
    
    for i in range(1, menor+1):
        v_max = maior*i
        if divide(maior, menor, v_max) == 1:
            print(f'O MMC entre {m} e {n} é {v_max}')
            break
    
main()
````

## Exercício 4
(a) Escreva uma função com protótipo
```C
void somabit (int b1, int b2, int *vaium, int *soma);
```
que recebe três bits (inteiros entre $0$ e $1$) $b1$, $b2$ e $*$vaium e devolve um bit soma representando a soma dos três e o novo um bit "vai-um" em *vaium.

(b) Escreva um programa que leia dois números em binário e calcula um número em binário que é a soma dos dois números dados. Utilize a função acima.

````python
def somabit(b1: int, b2: int, vaium: int) -> tuple[int, int]:

    som = b1 + b2 + vaium
    if som == 0:
        return 0, 0
    elif som == 1:
        return 1, 0
    elif som == 2:
        return 0, 1
    else:  # som == 3
        return 1, 1

def main():
    m = input('binário 1: ')
    n = input('binário 2: ')
    
    # Garantir que os números tenham o mesmo comprimento
    if len(m) > len(n):
        n = '0' * (len(m) - len(n)) + n
    else:
        m = '0' * (len(n) - len(m)) + m
    
    vaium = 0
    result = ''
    
    for i in range(len(m)-1, -1, -1):
        soma, vaium = somabit(int(m[i]), int(n[i]), vaium)
        result = str(soma) + result
    
    if vaium == 1:
        result = '1' + result
    
    print(f'soma: {result}')
    
main()
````

## Exercício 5
(a) Escreva uma função com o protótipo 
```C
void converte (char ch, int *tipo, char *valor);
```
que recebe um caractere ch e devolve em *tipo $0$, se o caractere for um número inteiro, $1$ se for uma letra (maiúscula ou minúscula) e $2$ caso contrário; e além disso, no caso de ser uma letra, converte para maiúscula, senão devolve ch inalterado.

(b) Escreva um programa que leia uma seqüência de $n$ caracteres e imprima a seqüência convertida para maiúscula, eliminando os caracteres que não forem letras ou números.


````python
def converte(ch:str) -> tuple[int, str]:
    num = '0123456789'
    alpha = 'qwertyuiopasdfghjklzxcvbnm'
    if ch in num:
        return 0, ch
    elif ch in alpha:
        return 1, ch.upper()
    else:
        return 2, ch
    
def main():
    n = int(input('n:'))
    seq = [input(f'{i+1}') for i in range(n)]
    
    for j in seq:
        ret, ch = converte(j)
        if ret == 0:
            print(ch)
        elif ret == 1:
            print(ch)

    
main()
````

(a) Só para ver se você entendeu as fórmulas, qual é o valor de $F_{6}$ e $G_{6}$?

(b) Faça uma função de nome *valor* que recebe um inteiro $k \geq 1$ e devolve $F_{k}$ e $G_{k}$.

  Exemplo: Para $k=2$, a função deve devolver os valores $1$ e $2$. Para $k=3$, a função deve devolver os valores $3$ e $8$. Para $k=4$, a função deve devolver os valores $8$ e $11$.

 (c) Faça um programa que lê um inteiro $n > 2$ e imprime os valores $F_{n-2} + G_{n+200}$ e $F_{n+200} - G_{n-2}$

Seu programa deve obrigatoriamente utilizar a função do item anterior, mesmo que você não a tenha feito.

````python
# (a) F_6 = 59, G_6 = 44

def valor(max_n: int) -> tuple[list[int], list[int]]:
    """
    Retorna duas listas contendo, respectivamente, os valores de F(1) até F(max_n) e G(1) até G(max_n).
    Assim, f[k] representará F(k) e g[k] representará G(k).
    """
    f = [0] * (max_n + 1)
    g = [0] * (max_n + 1)
    
    f[1] = 2
    f[2] = 1
    g[1] = 1
    g[2] = 2
    
    for k in range(3, max_n + 1):
        f[k] = 2 * f[k - 1] + g[k - 2]
        g[k] = g[k - 1] + 3 * f[k - 2]
    
    return f, g

def main():
    n = int(input("Digite um n > 2, n: "))
    
    if n < 3:
        print("n deve ser maior que 2.")
        return
    
    max_val = n + 200

    f, g = valor(max_val)
    
    print(f"F({6}) = {f[6]}")
    print(f"G({6}) = {g[6]}")
    
    # Calcula e exibe os resultados pedidos
    print(f"F({n-2}) + G({n+200}) = {f[n-2] + g[n+200]}")
    print(f"F({n+200}) - G({n-2}) = {f[n+200] - g[n-2]}")


main()
````
