## Exercício 1
Dada uma sequência de $n$ números, imprimi-la na ordem inversa à da leitura.

```python
def main():
    n = int(input('n: '))
    l = []
    
    for i in range(n):
        num = float(input(f'Digite o {i+1}º número: '))
        l.append(num)
    
    for i in range(1, len(l) + 1):
        print(f'{l[-i]} ', end='')
    
    
main()
```

## Exercício 2
Deseja-se publicar o número de acertos de cada aluno em uma prova em forma de testes. A prova consta de $30$ questões, cada uma com cinco alternativas identificadas por $A, B, C, D$ e $E$. Para isso são dados:
* O cartão gabarito;
* O número de alunos da turma;
* O cartão de respostas para cada aluno, contendo o seu número e suas respostas.

```python
def main():
    n_alunos = int(input('Digite o número de alunos: '))
    gabarito = []
    notas = []
    
    for i in range(1, 31):
        x = input(f'Digite o gabarito da {i}ª questão: ')
        gabarito.append(x)
        
    for n in range(n_alunos):
        acertos = 0
        prova_aluno = []
        for i in range(0, 30):
            x = input(f'Digite a {i+1}ª resposta do aluno {n +1}: ')
            prova_aluno.append(x)
            
            if prova_aluno[i].lower() == gabarito[i].lower():
                acertos += 1
        
        notas.append(acertos)
        
    for i in range(len(notas)):
        print(f'Aluno: {i+1} / Nota: {notas[i]}/30')
    
    
main()
```

## Exercício 3
Tentando descobrir se um dado era viciado, um dono de cassino honesto (ha! ha! ha! ha!) o lançou %n% vezes. Dados os $n$ resultados dos lançamentos, determinar o número de ocorrências de cada face.

## Exercício 4
Dados dois vetores $x$ e $y$, ambos com $n$ elementos, determinar o produto escalar desses vetores.

```python
def main():
    d = int(input('Digite a dimensão dos vetores: '))
    x = []
    y = []
    prod = 0
    
    for i in range(d):
        v = float(input(f'Digite o {i+1}º elemento do vetor x: '))
        x.append(v)
    
    for i in range(d):
        v = float(input(f'Digite o {i+1}º elemento do vetor y: '))
        y.append(v)
    
    for i in range(d):
        prod += (x[i]*y[i])
    
    print(f'O produto vetorial de x e y é {prod}')
    
main()    
```

## Exercício 6
*(COMP 89)* Dados dois strings (um contendo uma frase e outro contendo uma palavra), determine o número de vezes que a palavra ocorre na frase.

* Exemplo:

Para a palavra **ANA** e a frase :

**ANA** E MARI**ANA** GOSTAM DE B**ANANA**

Temos que a palavra ocorre $4$ vezes na frase.

```python
def main():
    frase = input('Insira uma frase: ')
    palavra = input('Insira uma palara para achar na frase: ')
    
    cont = 0
    
    for i in range(len(frase)):
        if palavra.lower() == frase[i:len(palavra) + i]:
            cont += 1
            
    print(f'Quantidade de vezes que a palavra {palavra} apareceu na frase: {cont}')
            
        
main()
```

## Exercício 7
*(MAT 88)* Dada uma seqüência de $n$ números reais, determinar os números que compõem a seqüência e o número de vezes que cada um deles ocorre na mesma.

* Exemplo: $n = 8$

Seqüência: $-1.7$,  $3.0$,  $0.0$,  $1.5$,  $0.0$, $-1.7$,  $2.3$, $-1,7$

* Saída:           
  * $-1.7$ ocorre $3$ vezes
  * $3.0$ ocorre $1$ vez
  * $0.0$ ocorre $2$ vezes
  * $1.5$ ocorre $1$ vez
  * $2.3$ ocorre $1$ vez


```python
def main():
    n = int(input('n: '))
    seq = []
    appear = []
    
    for i in range(n):
        x = float(input(f'{i+1}º: '))
        seq.append(x)
        
    for i in range(len(seq)):
        cont = 0
        if seq[i] not in appear:
            for j in range(len(seq)):
            
                if seq[i] == seq[j]:
                    cont +=1
            appear.append(seq[i])
            
            if cont == 1:
                print(f'{seq[i]} ocorre {cont} vez')
            else:
                print(f'{seq[i]} ocorre {cont} vezes')
        
    
main()
```

## Exercício 8
 Dados dois números naturais $m$ e $n$ e duas sequências ordenadas com $m$ e $n$ números inteiros, obter uma única seqüência ordenada contendo todos os elementos das seqüências originais sem repetição.

Sugestão: Imagine uma situação real, por exemplo, dois fichários de uma biblioteca.

```python
def main():
    m = int(input('m: '))
    n = int(input('n: '))
    list_m = []
    list_n = []
    listao = [] 
    listao_sorted = []
    
    for i in range(m):
        x = int(input(f'Digite o {i+1}º elemento da lista m: '))
        list_m.append(x)
        listao.append(x)
    
    for i in range(n):
        y = int(input(f'Digite o {i+1}º elemento da lista n: '))
        list_n.append(y)
        listao.append(y)
    
    for i in range(len(listao)):
        for index, valor in enumerate(listao_sorted):
            if listao[i] < valor:
                listao_sorted.insert(index, listao[i])
                break
        else:
            listao_sorted.append(listao[i])
    
    print(f'Lista M: {list_m}')
    print(f'Lista N: {list_n}')
    print(f'Lista com todos os elementos ordenados: {listao_sorted}')
    
    
main()
```

## Exercício 10

```python
def main():
    n = int(input('Digite o grau do polinômio: '))
    coef = []
    ponto = []
    
    print('p(x) = a0', end='')
    
    for i in range(1, n+1):
        print(f' + a{i}*x^{i}', end='')
        
    for i in range(n+1):
        x = float(input(f'\nDigite o valor do coeficiente a{i}: '))
        coef.append(x)
    
    print(f'\np(x) = {coef[0]}', end = '')
    for i in range(1, len(coef)):
        print(f' + {coef[i]}*x^{i}', end='')
        
    
    k = int(input('\nQuantos pontos deseja verificar o valor? '))
    
    if k >= 0:
        for i in range(k):
            x = float(input(f'x{i+1}: '))
            ponto.append(x)
        
        for i in range(len(ponto)):
            soma = coef[0]
            for j in range(1,len(coef)):
                soma += ponto[i]*coef[j]
            
            print(f'p({ponto[i]}) = {soma}')
        
    else:
        print('Bye Bye')
    
main()
```

## Exercício 15
Em uma classe há $n$ alunos, cada um dos quais realizou $k$ provas com pesos distintos. Dados $n$ , $k$, os pesos das $k$ provas e as notas de cada aluno, calcular a média ponderada das provas para cada aluno e a média aritmética da classe em cada uma das provas.


```python
def main():
    n = int(input('Digite o número de alunos: '))
    k = int(input('Digite o número de provas: '))
    pesos = [ float(input(f'Digite o peso da prova {i+1}: ')) for i in range(k)]  # list comprehensions
    medias = []
    soma_pesos = 0
    soma_medias = 0
    
    for i in range(len(pesos)):
        soma_pesos += pesos[i]
        
    for i in range(n):
        notas = []
        soma_notas = 0
        for j in range (k):
            x = float(input(f'Insira a nota da {j+1}ª prova do {i+1}º aluno: '))
            notas.append(x)
    
            soma_notas += notas[j]*pesos[j]
            
        medias.append(soma_notas/soma_pesos)
    
    for i in range(n):
        print(f'Média do aluno {i+1}: {medias[i]}')
        soma_medias += medias[i]    
            
    print(f'Média da turma: {soma_medias/n}')
    
    
main()
```

## Exercício 17
Dada uma seqüência de $n$ números inteiros, determinar um segmento de soma máxima.

Exemplo: Na seqüência $5, 2, -2, -7, 3, 14, 10, -3, 9, -6, 4, 1$ , a soma do segmento é $33$.

```python
def main():
    n = int(input('n: '))
    seq = [int(input(f'Digite o {i+1}º elemento da sequência: ')) for i in range(n)]  # list comprehensions
    soma_max = 0
    
    for i in range(n):
        for j in range(i+1, n):
            soma = 0
            for num in seq[i:j+1]:
                soma += num
            
            if soma > soma_max:
                soma_max = soma
        
    print(soma_max)
    
    
main()
```
